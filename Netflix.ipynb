#Data Cleaning

```python
df['country']=df['country'].fillna('Unkown')
```

```python
df['director']=df['director'].fillna('Unkown')
```

```python
df=df.dropna(subset=['date_added','rating'])
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df['date_added']=pd.to_datetime(df['date_added'])
```

```python
df
```

```python
df['Year_added']=df['date_added'].dt.year
```

```python
df
```

```python
df.info()
```

#Visualized

```python
sns.countplot(data=df,x="type",palette='deep',hue='type')
plt.title('Movies vs TV show')
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df,x='rating',palette='rocket',hue='rating')
plt.title('Who Can Watch')
plt.ylabel("")
plt.show()
```

```python
top_countries=df['country'].value_counts().head(10)
```

```python
top_countries
```

```python
sns.barplot(x=top_countries.values,y=top_countries.index,hue=top_countries,palette='rocket')
plt.title('Top 10 Countries by Contant')
plt.ylabel("")
plt.show()
```

```python
top_director=df['director'].value_counts().head(5)
```

```python
top_director
```

```python
sns.barplot(x=top_director.values,y=top_director.index,hue=top_director,palette='rocket')
plt.title('Top Dirctors')
plt.ylabel("")
plt.show()
```

```python
sns.histplot(data=df,x=df['release_year'])
plt.title("Release over Years Distribution ")
plt.ylabel('')
plt.show()
```

```python
sns.histplot(data=df,x=df['date_added'],bins=20)
plt.title("Year Added Distribution]")
plt.ylabel('')
plt.show()
```
